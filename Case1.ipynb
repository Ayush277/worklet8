 Bayesian Optimization for Hyperparameter Tuning


```python
from bayes_opt import BayesianOptimization
from sklearn.model_selection import cross_val_score, StratifiedKFold
from lightgbm import LGBMClassifier
import numpy as np
X_subset = X_new_train[:40000]
y_subset = y_train[:40000]

def lgbm_clf_bo(max_depth, num_leaves, learning_rate, min_child_weight, subsample, colsample_bytree):
    max_depth = int(round(max_depth))
    num_leaves = int(round((2 ** max_depth) * num_leaves))

    params = {
        'objective': 'binary',
        'random_state': 42,
        'boosting_type': 'gbdt',
        'metric': 'recall',
        'max_depth': max_depth,
        'num_leaves': num_leaves,
        'learning_rate': learning_rate,
        'min_child_weight': min_child_weight,
        'subsample': subsample,
        'colsample_bytree': colsample_bytree,
        'n_jobs': -1,
        'verbosity': -1
    }
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    score = cross_val_score(
        LGBMClassifier(**params), X_subset, y_subset, 
        scoring='recall', cv=cv, n_jobs=-1
    ).mean()

    return score
params_bounds = {
    'max_depth': (3, 10),
    'num_leaves': (0.1, 1), 
    'learning_rate': (0.01, 0.3),
    'min_child_weight': (1e-6, 1e-1),
    'subsample': (0.6, 1.0),
    'colsample_bytree': (0.6, 1.0)
}

lgbm_bo = BayesianOptimization(f=lgbm_clf_bo, pbounds=params_bounds, random_state=42, verbose=2)
lgbm_bo.maximize(init_points=3, n_iter=20)
```
