```python
unique_months = df['current_month'].unique()
unique_months = [m for m in unique_months if m != 0]
```

```python
m_cols = sorted([col for col in df.columns if col.startswith('m')],
                key=lambda x: int(x[1:]))

other_cols = [col for col in df.columns if not col.startswith('m')]
new_order = other_cols + m_cols
df = df[new_order]
```

```python
import numpy as np

m_cols = [f"m{i}" for i in range(2, 26)]

for idx, row in df.iterrows():
    current_month = row['current_month']
    keep_upto = current_month + 2 
    cols_to_empty = [f"m{i}" for i in range(keep_upto+1, 26)]
    for col in cols_to_empty:
        if col in df.columns:
            df.at[idx, col] = ''
```

```python
import pandas as pd

m_cols = [col for col in df.columns if col.startswith('m') and not col.endswith('_delay')]

new_order = []
for col in df.columns:
    new_order.append(col)
    if col in m_cols:
        delay_col = f"{col}_delay"
        if delay_col not in df.columns:
            df[delay_col] = '' 
        new_order.append(delay_col)
df = df[new_order]
```

```python
import numpy as np

m_cols = [col for col in df.columns if col.startswith('m') and not col.endswith('_delay')]

for m_col in m_cols:
    delay_col = f"{m_col}_delay"
    mask = df[m_col] == 0
    df.loc[mask, delay_col] = np.random.randint(5, 21, size=mask.sum())
```
