```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Define the function
def get_X_y_for_month(df, current_month):
    df = df.copy()

    demographic_features = [
        'interest_rate', 'unpaid_principal_bal', 'Loan_term', 'loan_to_value',
        'number_of_borrowers', 'debt_to_income_ratio', 'borrower_credit_score',
        'co-borrower_credit_score', 'insurance_percent', 'insurance_type',
        'loan_purpose', 'EducationLevel', 'MaritalStatus', 'Age',
        'Gender', 'EmploymentStatus', 'NumberOfDependents', 'Annual Income'
    ]

    month_features = [f"m{i}" for i in range(1, 8)]
    delay_features = [f"m{i}_delay" for i in range(1, 8)]
    features = demographic_features + month_features + delay_features

    target_col = f"m{current_month + 1}"
    for col in features:
        if col not in df.columns:
            df[col] = np.nan

    X = df[features]
    y = df[target_col] if target_col in df.columns else pd.Series([np.nan] * len(df), index=df.index)

    return X, y, features


```

```python
all_X_train, all_X_test = [], []
all_y_train, all_y_test = [], []

all_feature_set = set()

for month in sorted(data['current_month'].unique()):
    _, _, features = get_X_y_for_month(data[data['current_month'] == month], month)
    all_feature_set.update(features)

all_features = sorted(all_feature_set)

for month in sorted(data['current_month'].unique()):
    print(f"\nProcessing current_month = {month}")
    
    X, y, _ = get_X_y_for_month(data[data['current_month'] == month], current_month=month)

    for col in all_features:
        if col not in X.columns:
            X[col] = np.nan
    X = X[all_features]

    valid_idx = y.notna()
    X = X.loc[valid_idx]
    y = y.loc[valid_idx]

    if len(y) == 0:
        continue

    class_counts = y.value_counts()
    stratify_param = y if (class_counts >= 2).all() else None

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=12, stratify=stratify_param
    )

    all_X_train.append(X_train)
    all_X_test.append(X_test)
    all_y_train.append(y_train)
    all_y_test.append(y_test)

```

```python
from sklearn.metrics import precision_score, accuracy_score, recall_score, confusion_matrix

def train_classifier(clf, X_train, y_train, X_test, y_test, average='binary'):
    """
    Trains and evaluates a classifier.
    Returns accuracy, precision, recall, and specificity (for binary classification).
    """
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average=average, zero_division=0)
    recall = recall_score(y_test, y_pred, average=average, zero_division=0)
   
    if len(set(y_test)) == 2:
        tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0
    else:
        specificity = None  
    
    return accuracy, precision, recall, specificity

```

```python
def predict_rf_tabpfn(X, leaves, tabpfn_models, rf):
    preds = []
    for i, leaf in enumerate(leaves):
        # Ensure hashable key for the dictionary
        if isinstance(leaf, np.ndarray):
            hashable_leaf = tuple(leaf.tolist())
        else:
            hashable_leaf = leaf
        model = tabpfn_models.get(hashable_leaf)
        if model:
            pred = model.predict(X.iloc[[i]])
            preds.append(pred[0])
        else:
            pred = rf.predict(X.iloc[[i]])
            preds.append(pred[0])
    return np.array(preds)

y_pred = predict_rf_tabpfn(X_new_test, test_leaves, tabpfn_models, rf)
```

```python
tabpfn_models = {}
unique_leaves = np.unique(train_leaves)

for leaf in unique_leaves:
    idx = np.where(train_leaves == leaf)[0]
    if len(idx) > 10000:
        idx = np.random.choice(idx, 10000, replace=False)
    X_leaf = X_new_train.iloc[idx]
    y_leaf = y_train_final.iloc[idx]
    clf = TabPFNClassifier(device='cuda' if torch.cuda.is_available() else 'cpu')
    clf.fit(X_leaf.values, y_leaf.values)
    tabpfn_models[leaf] = clf
```

```python
def predict_rf_tabpfn(X, leaves, tabpfn_models, rf):
    preds = []
    for i, leaf in enumerate(leaves):
        if isinstance(leaf, np.ndarray):
            hashable_leaf = tuple(leaf.tolist())
        else:
            hashable_leaf = leaf
        model = tabpfn_models.get(hashable_leaf)
        if model:
            pred = model.predict(X.iloc[[i]])
            preds.append(pred[0])
        else:
            pred = rf.predict(X.iloc[[i]])
            preds.append(pred[0])
    return np.array(preds)
```

```python
y_pred = predict_rf_tabpfn(X_new_test, test_leaves, tabpfn_models, rf)

accuracy = accuracy_score(y_test_final, y_pred)
precision = precision_score(y_test_final, y_pred, zero_division=0)
recall = recall_score(y_test_final, y_pred, zero_division=0)
f1 = f1_score(y_test_final, y_pred, zero_division=0)

cm = confusion_matrix(y_test_final, y_pred)
tn, fp, fn, tp = cm.ravel()
specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0

print(f"Accuracy     : {accuracy:.4f}")
print(f"Precision    : {precision:.4f}")
print(f"Recall       : {recall:.4f}")
print(f"Specificity  : {specificity:.4f}")
print(f"F1 Score     : {f1:.4f}\n")

print("Classification Report:")
print(classification_report(y_test_final, y_pred, digits=4))
```
