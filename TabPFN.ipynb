```python
import numpy as np
def get_column_min_max(df, column_name):
    if column_name not in df.columns:
        return f"Column '{column_name}' does not exist in the DataFrame."

    if not np.issubdtype(df[column_name].dtype, np.number):
        return f"Column '{column_name}' is not numeric."

    max_val = df[column_name].max()
    min_val = df[column_name].min()
    return max_val, min_val


column_name = "Annual Income"
max_val, min_val = get_column_min_max(df, column_name)
print(f"Max of {column_name}:", max_val)
print(f"Min of {column_name}:", min_val)
```
